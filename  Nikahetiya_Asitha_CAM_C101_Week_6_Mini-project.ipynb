### Useful Function Definitions

```python
#create boxplots for dataframe (5 features)
def create_boxplots(df):
  fig, axes = plt.subplots(1, 5, figsize=(25, 5))

  # Flatten the axes array
  axes = axes.flatten()

  # Iterate through columns and create box plots
  for i, col in enumerate(df.columns):
      sns.boxplot(ax=axes[i], y=df[col])
      axes[i].set_title(col)

  plt.tight_layout()
  plt.show()

#plot histogram for dataframe (5 features)
def plot_histogram(df):

  fig, axes = plt.subplots(1, 5, figsize=(25, 5))

  # Flatten the axes array
  axes = axes.flatten()

  # Iterate through columns and create box plots
  for i, col in enumerate(df):
      sns.histplot(ax=axes[i], x=df[col])
      axes[i].set_title(col)

  plt.tight_layout()
  plt.show()

#print outlier proportions based on IQR
def check_outliers(df):
  n = df.shape[0]
  for col in df.columns:
      Q1 = df[col].quantile(0.25)
      Q3 = df[col].quantile(0.75)
      IQR = Q3 - Q1
      lower_bound = Q1 - 1.5 * IQR
      upper_bound = Q3 + 1.5 * IQR
      outlier_count = ((df[col] < lower_bound) |
      (df[col] > upper_bound)).sum()
      proportion_outliers = outlier_count / n
      print(f"Proportion of outliers in {col}: {proportion_outliers:.1%}")

#extract iqr bounds
def iqr_bounds(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound
```

### Cluster number, k, selection

#### Elbow method

```python
# Elbow method: plot WCSS for different k
wcss=[]

for i in range(2,15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    #kmeans.fit(X_scaled)
    cluster_labels = kmeans.fit_predict(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2,15),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

**Observation:** A bend occurs at 6 making it the most likely choice for k


#### Silhouette method

```python
#take a smaller sample to reduce computation time
X_scaled_sample = X_scaled.sample(frac=0.3, random_state=42)

for n_clusters in range(2,10):
  kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42, n_init='auto')
  cluster_labels=kmeans.fit_predict(X_scaled_sample)

  silhouette_avg = silhouette_score(X_scaled_sample, cluster_labels)
  print(f"For {n_clusters} clusters the silhouette score is {silhouette_avg}")

  fig, (ax1) = plt.subplots(1, 1)

  # The silhouette coefficient can range from -1, 1
  ax1.set_xlim([-1, 1])
  # The (n_clusters+1)*10 is for inserting blank space between silhouette
  # plots of individual clusters, to demarcate them clearly.
  ax1.set_ylim([0, len(X_scaled_sample) + (n_clusters + 1) * 10])

  # Compute the silhouette scores for each sample
  sample_silhouette_values = silhouette_samples(X_scaled_sample, cluster_labels)

  y_lower = 10
  for i in range(n_clusters):
      # Aggregate the silhouette scores for samples belonging to
      # cluster i, and sort them.
      ith_cluster_silhouette_values = \
          sample_silhouette_values[cluster_labels == i]

      ith_cluster_silhouette_values.sort()

      size_cluster_i = ith_cluster_silhouette_values.shape[0]
      y_upper = y_lower + size_cluster_i

      color = cm.nipy_spectral(float(i) / n_clusters)
      ax1.fill_betweenx(np.arange(y_lower, y_upper),
                        0, ith_cluster_silhouette_values,
                        facecolor=color, edgecolor=color, alpha=0.7)

      # Label the silhouette plots with their cluster numbers at the middle.
      ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

      # Compute the new y_lower for next plot.
      y_lower = y_upper + 15

  ax1.set_xlabel("The silhouette coefficient values")
  ax1.set_ylabel("Cluster label")

  # The vertical line for average silhouette score of all the values.
  ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

  ax1.set_yticks([])  # Clear the yaxis labels / ticks.
  ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

  plt.suptitle(("Silhouette plot for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

**Observation:** For all considered k, there are negative silhouette coeffient values suggesting less than optimal clustering. Silhouette score drops significantly after k=6. From the spectral plot, k=6 seems to offer the most even splits among clusters.

#### Hierarchical clustering


**Comparing linkage methods & distance metrics on a very small sample to pick best one**

```python
import itertools
#take a very small sample to reduce computation time
tiny_X = X_scaled.sample(frac=0.01, random_state=42)

# Function to create a dendrogram using a given linkage metric
def plot_dendrogram(X, linkage_method, metric, ax):
    linked = linkage(X, metric=metric, method=linkage_method)
    dendrogram(linked, ax=ax)
    ax.set_title(f'Dendrogram ({linkage_method}-{metric}) Linkage)', fontsize=14)
    ax.set_xlabel('Sample index or (Cluster size)', fontsize=12)
    ax.set_ylabel('Distance', fontsize=12)

# Define the linkage methods
linkage_methods = [ 'complete', 'average']
# Define the metrics
metrics = ['euclidean', 'cityblock']

# Create a new plotting setup with the correct indexing
fig, axes = plt.subplots(1, len(linkage_methods)*len(metrics),figsize=(len(linkage_methods)*len(metrics)*10, 10))
#figsize=(len(linkage_methods)*len(metrics)*15, 10)

combinations = list(itertools.product(linkage_methods, metrics))

# Plot dendrograms for each linkage_method - metric combination
for i, (linkage_method, metric) in enumerate(combinations):
    plot_dendrogram(tiny_X, linkage_method, metric, axes[i])

# Display the plots
#plt.tight_layout()
plt.show()
```

**Observation:** *Complete* linkage with *manhattan* distance seems to give the most evenly spread out hierarchical clustering

### Visualization in 2-D

```python
def plot_2D_scatterplot(df, clusters, title, palette='viridis'):
  '''
  df: [nx2] dataframe containing 2D components.
  clusters: n-dimensional arry of cluster number
  '''
  df_plot = pd.DataFrame({
      'Component 1': df[:, 0],
      'Component 2': df[:, 1],
      'Cluster': clusters
  })

  #print("Sample df_plot:") # added print statement
  #print(df_plot.head(50))


  # Create scatterplot
  plt.figure(figsize=(10, 8))
  sns.scatterplot(x='Component 1', y='Component 2', hue='Cluster', palette=palette, data=df_plot)

  plt.xlabel('Component 1')
  plt.ylabel('Component 2')
  plt.title(title)
  plt.grid(True)
  plt.show()
```
